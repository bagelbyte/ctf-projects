### but first lets look at why we decided to look into this `code == GOAL`

why are we only looking into `code == GOAL` instead of considering other venues for getting a flag next? we assume, via the location of where user input is placed, that we will not be able to influence the execution of the program by changing the content of the instructions that live in the `.text` region of the program. because of this, we have a set collection of instructions, and the method for influencing the program is via changing the factors which are checked by the program during execution that would influence the path that the program takes. after our opportunity for input, we have the following

so we have the desire to make something in the program flow afterwards give us the flag, based on the input we give to the program. trying to make `code == GOAL` is one method of getting our flag (and it might be an easier method, considering making it to this branch will lead to the execution of `cat flag.txt`), but if that does not work we might have to consider doing something else and hope that the second code segment leads to codepaths that we can influence to produce a similar result (of the effect of `cat flag.txt`). so we have identified one viable codepath to achive our needs, we just need to figure out if its possible to execute along that codepath

### alright lets get back to it
we next should consider the positions of `code` and `clutter` in memory, and see how they relate to eachother. we suspect due to the ordering in the source file that `code` is at a higher memory address than `clutter`, but we will need to verify that. the easiest way of checking this will probably be in ghidra, so lets spin that up. 



we see here that the memory for the character array (`clutter`) is going to start at `EBP - 0x110`, and `code` will start at `EBP - 0x8` 
> note that the stack addresses in pink are off by 8 bytes, because of the base pointer from the previous frame. these addresses are listed as offsets from the previous frame, but you see the right values lower in the assembly

since we know that the `clutter` memory region starts at `EBP - 0x110`, and writing to memory starts from low to high, the first `0x110 - 0x8 = 0x108` bytes will be a part of the `clutter` memory space while the next `0x8` bytes written will write to the `code` memory region. because we know that we can write whatever we want, we just want to write <`0x108` bytes> followed by 8 bytes. these remaining 8 bytes, when interpreted as a `long` type, should be equivalent to the hex number `0xdeadbeef`.  

### crafting the contents of `code`

the next thing that we want to do is figure out what bitsequence `x` we need to write to the memory region of `code`, such that `(8 byte long)x == 0xdeadbeef`. important here is to figure out the way in which the contents of memory is interpreted. the most important factor in this case is endianness, which determines how numerical values are encoded into bitstrings. 



we see that the binary uses little endian, so we need to make sure that our value of `0xdeadbeef` is encoded in little endian so that it can later be decoded in little endian. 

````python
import pwn
code_value_little_endian_bytes = pwn.p64(0xdeadbeef, endian='little')
````

````python
code_value_little_endian_bytes
````

and now we just need to pad this from the front with `0x108` bytes to fill up `clutter`. well we know each ascii character is going to be 8 bits, so hopefully if we send an ascii encoded letter "a", that we will get us 1 byte written. do this `0x108` times and we get

````python
clutter_buffer_string = "a" * 0x108
payload = (clutter_buffer_string).encode("ascii") + code_value_little_endian_bytes + "\n".encode("ascii") 
payload
````

So from what we have statically deduced, we know that this SHOULD overwrite the memory region associated with the variable `code`, and probably give us back the path we want. lets just send it

````python
conn = pwn.remote("mars.picoctf.net", 31890)
print(conn.clean().decode('ascii'))
````

````python
conn.send(payload)
print(conn.clean().decode('ascii'))
````
