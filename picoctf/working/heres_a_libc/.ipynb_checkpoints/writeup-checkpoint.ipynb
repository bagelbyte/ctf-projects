Next I want to figure out why Codeql did not find that, because it seems like a fairly common occurance to look for instances where unconstrained user input is placed into some buffer. In running more of the codeql test suites (cpp-lgtm-full.qls, cpp-code-scanning.qls, cpp-security-extended.qls, cpp-security-and-quality.qls), there are also no detected buffer overflows, even though there are a few items in the test suites that handle looking for buffer overflows.

Of particular interest is the function that is supposed to handle catching unsafe `scanf` usage, found at `cpp/ql/src/experimental/Security/CWE/CWE-120/MemoryUnsafeFunctionScan.ql`. 

```
/**
 * @name Scanf function without a specified length
 * @description Use of one of the scanf functions without a specified length.
 * @kind problem
 * @problem.severity warning
 * @id cpp/memory-unsafe-function-scan
 * @tags reliability
 *       security
 *       external/cwe/cwe-120
 */

import cpp
import semmle.code.cpp.commons.Scanf

from FunctionCall call, ScanfFunction sff
where
  call.getTarget() = sff and
  call.getArgument(sff.getFormatParameterIndex()).getValue().regexpMatch(".*%l?s.*")
select call, "Dangerous use of one of the scanf functions"

```

this regex string matches on items that contain unbounded strings, but dont handle unbounded [scansets](https://www.cplusplus.com/reference/cstdio/scanf/). the fix is simple, to include the scanset in the regex. so `".*%l?s.*"` becomes `.*%l?(s|\\[(?s:.)*\\]).*` 
> the regex used by codeql is java regex, so `\\[` is equivalent to the literal `[`. also we have to ensure to allow for newline characters in the regex, so we add the `(?s:.)*` instead of just `.*` inside the brackets

now if we rerun the query, we find the vulnerable instance

#### 3.3 the implication

what is the implication of this vulnerability? when we find something vulnerable, usually we hope it means that we can make something bad happen with it. however, the bad things we can make happen are reliant what values are used in subsequent steps of the program's execution. taking a look at the initial scope of impact of this vulnerability, we know that we can write an arbitrary amount of data, starting from a low address, up to higher memory addresses. according to ghidra (from low to high memory addresses) on the stack frame for `do_stuff` we will have the `1ocal_88` variable, followed by `local_89`, followed by `uVar1`. if we pass this, we begin to overwite data that supports the stack frame itself, such as the return address to the previous execution path. 

>for the remainder of this writeup, I am going to be using Joern because it works out better for analyzing decompiled code. I'll probably come back to CodeQL whenever we go back to a program that we have the source for

So, we have the ASSUMPTION that if there is a way for us to get a shell, there will be something that we can overwrite that will help us get to the point in which a shell will be executed (its at least a single step, if not the only step needed)

considering the location of the scanf location on the stack, we have the following order of values that we will write over (on the stack)

local_88 (112 bytes) -> local_18 (8 bytes) -> local_10 (8 bytes) -> base pointer -> return address -> caller stack data

because any one of these things can be overwritten, we really just need to find something interesting that we want to overwrite that would lead to us being able to call "system" or the like. put another way, let the set of (data) structures that are subsequently read from be denoted as the set $D_R$, and structures that can be written to as $D_W$. assuming any computing instruction is done as a series of reads and writes, and lets say that we want a given structure to have some value `x`. assuming that the intial execution of the program with NO input gives us the thing that we want, we basically are looking for the intersection of $D_R$ and $D_W$ so that there is some sequence of "executions" from a structure $ x \in D_R \cap D_W$ that lead to the state that we care about. 

one way to think about these values is by considering the perspective that to "write" a value, you must first "read" a value from somewhere. then we might consider the intersection $D_R \cap D_W$, where $D_W$ is comprised of the things we can write to in the `scanf` operation, and $D_R$ are subsequent reads, leading to $D_R \cap D_W$ comprising of subsequent reads from our writable "surface"


##### the desired state

we have a few states that we may want, but lets first consider the state of getting the libc `system` function to run. `system` takes in a single string as an argument, and "ls -l" will suffice for now to prove if we were successful.

##### solution complexity
do we just try out arbitrary modifications to each of the structures in our intersection? imo the difficuluty in figuring out what to do next revolves around the thought that we have some end state we want to reach, and there are a thousand gremlins influencing what happens between any two points in the program, and we have to reason about all of those in betweens to figure out what the program will do. however, we might at any point consider what the smallest amount of reasoning is necessary to evalue an option for its value in achieving a goal. 

remember that we showed that we can overwrite `local_88 (112 bytes) -> local_18 (8 bytes) -> local_10 (8 bytes) -> base pointer -> return address -> caller stack data` with our `scanf` function.

this gives us the following elements for $D_W$:
* local_88 (112 bytes) 
* local_18 (8 bytes)
* local_10 (8 bytes)
* base pointer
* return address 
* (caller stack data)*

> for the `caller stack data`, notice that we have it in a `()*`. I use this notation to denote that we have it as a blob of memory that we can further refine if we need to, but for now we will just denote it as we have 

we now begin considering data segments for usage in our end goal of calling `system`. we know that if we can overwrite the return address, we automatically gain the ability to abritrarily run whatver instructions, so we have the ability to move to the `system` function. however whats missing is the setup to the function, namely the passing of the address of the string that we want system to call in RDI (`system` only takes one argument, so we know the first argument is going to be in RDI). 

we can now divide the things we know into two sections in the context of its ability to execute `system`

* return address: 
  * known: can execute system
  * unknown: can RDI be populated with address of string prior to execution
* local_88, local_18, local_10, base pointer, (caller stack data)*: 
  * known: n/a
  * unknown: will some value lead to a "valid" (with proper RDI value) execution of system

 
i dont really have any reason to belive that evaluating implications of values on other variables will be more likely to produce the `system` execution that we want, so we will go ahead with the evaluation of the return address path. 

> some quick thoughts on why we dont feel like evaluating what RDI would be at the end of this function (to see if we can make it so RDI already has a value that we care for at the end of the function): in my head, the value of the control of the return address should first be looked at as the ability to control the next steps, so we should try and figure out a good sequence of "next steps" that get us to where we want to go. also, because we never really evaluated the impact of our input on the other data sections, i dont care to look at them yet for this use. 

what we want to find is a sequence of instructions that we can use to feed RDI with the value we want, then call `system`. 

##### context

registers, along with the stack values, are structures which give "context" to instructions. if we have a series of instructions that when run twice, produce the same result when run twice, then their context during both executions should be the same. as an example, if we do the series of instructions

z = x + y
b = z + y

twice, and we want the results (z and b in this case) to be the same after sequentially evaluating both statements, then to make this happen, during both evaluations, the values for "x" and "y" should be the same in both iterations.

the reason i bring this up is that if we want to run some instructions and produce consistient results during 2 program executions, then we need to run the same set of instructions with the same context during two executions. 



as far as i can tell, in this program the text sections (the things that get executed as instructions) stay consistent in their location, HOWEVER the stack location moves around. this makes it hard (impossible?) to give consistient context to any given set of instructions, assuming we can control the next instruction. if our stack frame locations were to stay constant, we would be able to overwrite the base pointer stored on the stack, giving the new instructions that we want to execute a context that we define. however because we dont know where the stack addresses will be on any given instruction, and if we were to overwrite the instruction pointer, we would be forced to give a value to overwrite the base pointer (on the stack). the best option we have is to set this address to be the address of some `text` segment, but it might be difficult to rely on these values to give us the exact stack context we want. so in this case, where we dont have a good way of defining our own context, we have to find instructions that dont rely on any context to execute the instructions we want (this would involve finding instructions that "define" the context, rather than executing "on" a context). 


##### but then a flash of inspiration (WILL IT WORK!?)

SO, the previous issue we had was that our stack was floating around in some memory region that we were writing to, and we had no clue where that was. it made it so that when we wanted to write context to the memory region where our stack was, we had no clue where we were writing to, making it impossible to provide what we just wrote as a context to some other region. this issue stemmed from us being given a (random?) context location. but what if we can move the context location to somewhere writable, then create ourselves some context, and then execute based off of that context? the hunt begins for a memory region, whose addresses are consistent, and has both read and write permissions. 



looking at the segment information, we see that there is a section mapped into memory at address `0x0000000000600e00`, 



to see if this works, the plan is to overwrite the base pointer with an address in the `00601000-00602000` page. we will use a value after the `.bss` segment, which ends at `0x601060` (aka we are going to use `0x601060` as our new "stacks" stack pointer (bottom of the stack)). because of the instructions we plan to run next, we want to offset this address with the same size as our `do_stuff`'s functions stack, which is `0x81`, along with 2 8 byte pointer values that we plan to shove in there, giving us a base pointer of `0x6010f1`

once we have moved our stack context over to that region of memory, we now know where our stack is, so we can repeat the same sequence of instructions that we just ran, which are instructions that start at the address `0x004006e3`.

alright so now we have what the context we would like those instructions to operate under. how do we get that to happen? well luckily, when we return from a given stack frame, we get to move to a context of a stack frame that lives above the current stack frame in memory. couple this with our ability to write above our current stack frame means that we can write a new stack frame (the one we just defined) for ourselves. so heres the plan - we know that we have

`local_88 (112 bytes) -> local_18 (8 bytes) -> local_10 (8 bytes) -> base pointer -> return address -> caller stack data`

and because we dont really want to deal with thinking about the implications of putting our data for the next stack frame in the current (`do_stuff`) frame, we are just going to have something that write something like this 
(local_88)                                                                                                               (local_10)
|                                                                                                                        |
QQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQ\0QQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQQ8888888800000000<new bp><new return address>
                                                                                                                 |              | 
                                                                                                                 (local_18)     (`do_stuff` RBP)

what we want is for our "new bp" to refer to the new stack frame that we created. because we know RBP is at `0x6010f1`, then our new RBP (for the crafted frame) should be `0x6010f1 + 8 + 8 + 0x80` (8 for the <new bp>, 8 for the <new return address>, and 0x80 for the new string payload), giving us a final value of `0x601181` for the value of <new bp> (this is importantly still in our original page) 

our <new return address> is going to be the location of the instructions we want to run, which is `0x400762`. putting it all together (remember little endian), we have
