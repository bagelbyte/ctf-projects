```C
  local_78 = 0x20656d6f636c6557;
  local_70 = 0x636520796d206f74;
  local_68 = 0x6576726573206f68;
  local_60 = 0x2172;
``` 

these lines also look suspiciously like ASCII. putting it into cyber chef and swapping the endianness, we get 



which is good, but its still not the flag. 

digging around in the `libc.so.6`, we also dont really see anything interesting... which probably means that the flag is not in the binary, and instead we have to get access to the flag somewhere outside of the binary's memory space (aka **its probably in a flat file on the host OS**).


### step 3 - find a bug that gives us a shell or something

at this point, we know that the program is vulnerable to SOMETHING (beacause it wouldnt be much of a ctf otherwise)... but we dont know what it is vulnerable to. but first we probably want to get the program able to run locally, because as it is things dont work




#### 3.1 - make vuln run again




#### 3.2 - finding vulnerable code paths

i've discovered CodeQL recently and think its pretty interesting, so I figure i'll give it a shot on this challenge. Codeql requires C source to be able to do its analysis, so we will see how far we get with ghidra's decompiler and parsing that into Codeql. After exporting the functions mentioned in the first section, we do string replacements for ghidra not knowing how to handle variables (see src/clean for what we do), which gives us `vuln_replaced.c`. Then we add some headers, giving us `vuln_todo.c`. This compiles, and we are able to build a Codeql database from this source. to begin with, i'm going to use the `Critial` test suite to scan for anything it thinks might be bad. 



Unfortunately, it doesn't see anything interesting, but lets do some static analysis ourselves. 



we see that in the `do_stuff` function, there are opportunities for user input at `scanf`. In the first `scanf` statement, the `%[^\n]` scanf  format string allows an unlimited amount of input, which can be written into `local_88`, as long as its not a newline character (`\n`). So we see a potential for a buffer overflow. This is confirmed when testing against the `vuln_patched` 



So we know that the buffer overflow exists. Next I want to figure out why Codeql did not find that, because it seems like a fairly common occurance to look for instances where unconstrained user input is placed into some buffer. In running more of the codeql test suites (cpp-lgtm-full.qls, cpp-code-scanning.qls, cpp-security-extended.qls, cpp-security-and-quality.qls), there are also no detected buffer overflows, even though there are a few items in the test suites that handle looking for buffer overflows.

Of particular interest is the function that is supposed to handle catching unsafe `scanf` usage, found at `cpp/ql/src/experimental/Security/CWE/CWE-120/MemoryUnsafeFunctionScan.ql`. 

```
/**
 * @name Scanf function without a specified length
 * @description Use of one of the scanf functions without a specified length.
 * @kind problem
 * @problem.severity warning
 * @id cpp/memory-unsafe-function-scan
 * @tags reliability
 *       security
 *       external/cwe/cwe-120
 */

import cpp
import semmle.code.cpp.commons.Scanf

from FunctionCall call, ScanfFunction sff
where
  call.getTarget() = sff and
  call.getArgument(sff.getFormatParameterIndex()).getValue().regexpMatch(".*%l?s.*")
select call, "Dangerous use of one of the scanf functions"

```

this regex string matches on items that contain unbounded strings, but dont handle unbounded [scansets](https://www.cplusplus.com/reference/cstdio/scanf/). the fix is simple, to include the scanset in the regex. so `".*%l?s.*"` becomes `.*%l?(s|\\[(?s:.)*\\]).*` 
> the regex used by codeql is java regex, so `\\[` is equivalent to the literal `[`. also we have to ensure to allow for newline characters in the regex, so we add the `(?s:.)*` instead of just `.*` inside the brackets

now if we rerun the query, we find the vulnerable instance



#### 3.3 exploiting the vulnerable code path
